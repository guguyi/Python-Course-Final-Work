```python
def read_list_file(filename):
    """
        è¯»å–æ–‡ä»¶ä¸­å†…å®¹ä¸ºåˆ—è¡¨çš„txtæ–‡ä»¶(æ¯è¡Œä»¥'\n'åˆ†éš”)
        å¦‚: åœç”¨è¯ï¼Œå¦å®šè¯ï¼Œç¨‹åº¦å‰¯è¯
        filename: æ–‡ä»¶è·¯å¾„
        return:file_list type: list
    """
    file_list = [k.strip() for k in open(
        filename, encoding='utf-8').readlines() if k.strip() != '\n']
    return file_list
```

```python
def read_dict_file(filename):
    """
        è¯»å–æ–‡ä»¶ä¸­å†…å®¹ä¸ºé”®å€¼å¯¹çš„txtæ–‡ä»¶
        å¦‚: BosonNLPæƒ…æ„Ÿè¯å…¸
        filename: æ–‡ä»¶è·¯å¾„
        return: classify_dict tyoe: dict
    """
    classify_list = [k.strip() for k in open(
        filename, encoding='utf-8').readlines() if k.strip() != '\n']
    classify_dict = dict()
    for i in classify_list:
        if len(i.split(' ')) == 2:
            classify_dict[i.split(' ')[0]] = i.split(' ')[1]
    return classify_dict
```

```python
def jieba_cut(string, stopwords, emoji_dict):
    """
        å¯¹æ¯ä¸€æ¡å¾®åšè¿›è¡Œjiebaåˆ†è¯å¹¶å»é™¤åœç”¨è¯
        string: å¾®åš type: str
        stopwords: åœç”¨è¯åˆ—è¡¨ type: list
        emoji_dict: emojiè¡¨æƒ…ä¸­æ–‡å¯¹ç…§
        return: å»é™¤åœç”¨è¯ã€ç¿»è¯‘emojiåçš„åˆ†è¯åˆ—è¡¨
    """
    string = string.replace(u'\xa0', u' ')  # å°†\xa0å­—ç¬¦æ›¿æ¢ä¸º' '
    word_list = jieba.lcut(string)
    word_list = [i for i in word_list if i not in stopwords and i != ' ']
    for i in range(0,len(word_list)):
        if word_list[i] in emoji_dict.keys():
            word_list[i] = emoji_dict[i]
    return word_list
```

```python
def senti_ana(string, stopwords, BosonNLP_dict, Nega_list, degree_dict):
    """
        å¯¹æ–‡æœ¬å†…å®¹è¿›è¡Œæƒ…æ„Ÿåˆ†æ, è¿”å›æƒ…æ„Ÿå¾—åˆ†
        string: å¾®åšæ–‡æœ¬å†…å®¹ str
        stopwords: åœç”¨è¯ list
        BosonNLP_dict: æƒ…æ„Ÿè¯å…¸ dict
        Nega_list: å¦å®šè¯è¯å…¸ list
        degree_dict: ç¨‹åº¦å‰¯è¯è¯å…¸ dict
        return: score æƒ…æ„Ÿå¾—åˆ†
    """
    word_list = jieba_cut(string, stopwords, emoji_dict)  # å¯¹æ–‡æœ¬å†…å®¹è¿›è¡Œåˆ†è¯

    sen_word = dict()  # æƒ…æ„Ÿè¯è¯å…¸ ç´¢å¼•ä¸ºåˆ†è¯ç»“æœåœ¨word_listä¸­çš„ä¸‹æ ‡
    nega_word = dict()  # å¦å®šè¯è¯å…¸ åŒä¸Š
    degree_word = dict()  # ç¨‹åº¦å‰¯è¯è¯å…¸ åŒä¸Š

    # å¯¹åˆ†è¯ç»“æœè¿›è¡Œåˆ†ç±»
    for i in range(len(word_list)):
        word = word_list[i]
        if word in BosonNLP_dict and word not in Nega_list and word not in degree_dict.keys():
            sen_word[i] = BosonNLP_dict[word]
        elif word in Nega_list and word not in degree_dict.keys():
            nega_word[i] = -1
        elif word in degree_dict.keys():
            degree_word[i] = degree_dict[word]

    # è®¡ç®—å¾—åˆ†
    score = 0  # åˆå§‹å¾—åˆ†ä¸º0
    weight = 1  # åˆå§‹æƒé‡ä¸º1
    sen_index = -1  # åˆå§‹æƒ…æ„Ÿè¯ç´¢å¼•ä¸º-1
    sen_index_list = list(sen_word.keys())  # æƒ…æ„Ÿè¯ç´¢å¼•åˆ—è¡¨

    for i in range(len(word_list)):
        if i in sen_word.keys():
            score += weight*float(sen_word[i])
            sen_index += 1
            if sen_index < len(sen_index_list)-1:
                for j in range(sen_index_list[sen_index], sen_index_list[sen_index+1]):
                    # å¦‚æœä¸¤ç›¸é‚»æƒ…æ„Ÿè¯ä¹‹é—´æœ‰å¦å®šè¯
                    if j in nega_word.keys():
                        weight = -weight
                    # å¦‚æœä¸¤ç›¸é‚»æƒ…æ„Ÿè¯ä¹‹é—´æœ‰ç¨‹åº¦å‰¯è¯
                    elif j in degree_word.keys():
                        weight *= float(degree_word[j])
            if sen_index < len(sen_index_list)-1:
                i = sen_index_list[sen_index+1]

    return score  # è¿”å›æƒ…æ„Ÿå¾—åˆ†
```

```python
all_filename = 'all_with_score.csv'

def write_csv(filename):
    try:
        result_headers = ['å¾®åšid', 'å¾®åšæ­£æ–‡/è½¬å‘ç†ç”±', 'æ˜¯å¦åŸåˆ›', 'æƒ…æ„Ÿå¾—åˆ†']
        with open('yyqx_csv.csv', 'r', encoding='utf-8') as f:
            reader=csv.reader(f)
            content = [row[:]for row in reader][1:]
        
        for i in range(0,len(content)):
            content[i].append(scorelist[i])

        with open(filename, 'a', encoding='utf-8-sig', newline='') as f:
            writer = csv.writer(f)
            writer.writerows([result_headers])
            writer.writerows(content)
        
    except Exception as e:
        print(e)
```
